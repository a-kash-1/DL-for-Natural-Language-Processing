```
import numpy as np
import cv2
```

```
image = cv2.imread('rectangle.jpg')
output = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```
# detect circles in the image
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, 200, 50)

# at least 1 circle is there
if circles is not None:
  # convert the x,y coordinates and radius of circle to integer
  circles = np.round(circles[0,:]).astype('int')

  # loop over the (x,y) coordinates and radius of the circle
  for (x,y,r) in circles:
    # draw the circle in the output image, then draw a rectangle
    # corresponding to the center of the circle
    cv2.circle(output, (x,y), r, (0,255,0), 4)
    cv2.rectangle(output, (x-5, y-5), (x+5, y+5), (0, 128, 255), -1)
```
